**2.** Пусть есть выборка $X_1, ..., X_n$. Опишите процедуру бутстрепа построения доверительного интервала для величины $\theta = \mathsf{E} X_1$. Рассмотрите три типа  бутстрепных доверительных интервалов.
Выпишите соответствующие формулы. 

**Описание**:

1. Производим случайную выборку с элементами из исходной выборки размера m. Таким образом, создаём начальную выборку.
2. Рассчитываем интересующую статистику по бутстрепной выборке: $\hat{\theta}$.
3. Проделаем первые 2 пункта B раз.
4. Строим доверительный интервал(ДИ), используя одну из формул ниже:
$$[2*\hat{\theta}-\hat{q}_{1-\frac{\alpha}{2}}; 2*\hat{\theta}+\hat{q}_\frac{\alpha}{2}]- центральный \ ДИ$$
$$[\hat{q}_\frac{\alpha}{2};\hat{q}_{1-\frac{\alpha}{2}}]-перцентильный \ ДИ$$ 
$$[\hat{\theta}+\hat{\sigma} \cdot z_{\frac{1+\alpha}{2}}; \hat{\theta}-\hat{\sigma} \cdot z_{\frac{1+\alpha}{2}}] -нормальный \ ДИ$$

**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

Для ускорения расчетов можете использовать broadcasting (не является обязательным требованием). Учите, что в таком случае внутри функции у вас могут получиться очень большие матрицы, из-за которых может кончиться оперативная память. 

```python
def bootstrap(sample, B=100000): 
    '''
    Считает бутстрепные оценки для исходной выборки

    :param sample: исходная выборка
    :param B: количество бутстрепных 
    :return bootstrap_estimations: оценки по бутстрепным выборкам
    '''
    sample = np.array(sample.dropna())
    bootstrap_estimations = []
    n= len(sample)
    for b in range(B):   
        sum_bsample = 0
        for i in range(n):
            # генерируем элемент бутстрепной выборки
            ind = np.random.randint(0, n)
            # добавляем его к сумме по этой выборке
            sum_bsample += sample[ind]
        current_mean = sum_bsample/n
        bootstrap_estimations.append(current_mean)

    return bootstrap_estimations
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. 

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу нормального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''
    bootstrap_estimations= np.array(bootstrap_estimations)
    v_boot= 1/len(bootstrap_estimations)*np.sum(bootstrap_estimations**2)-(1/len(bootstrap_estimations)*np.sum(bootstrap_estimations))**2
    z = sps.norm.ppf((1 - alpha) / 2)
    left, right = theta - z * v_boot**0.5, theta + z * v_boot**0.5

    return left, right

    
def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу центрального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    left, right= 2 * theta - np.nanquantile(bootstrap_estimations, [(1 - alpha) / 2,  alpha / 2])



    return left, right

    
def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу квантильного бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''   
    
    left, right = np.nanquantile(bootstrap_estimations, [alpha / 2, (1 - alpha) / 2])



    return left, right
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

```python
df = df[['ITSN1_N', 'DYRK1A_N', 'ELK_N']]
```

```python
dat = pd.DataFrame(index = ['normal', 'central', 'quantile', 'theta', 'interval_length'], columns=df.columns)
```

```python
for param in params:
    est = bootstrap(df[param])
    dat.loc['theta', param]= np.mean(df[param].dropna())
    dat.loc['normal', param] = bootstrap_normal_confidence_interval(np.mean(df[param]), est)
    dat.loc['central', param]= bootstrap_central_confidence_interval(np.mean(df[param]), est)
    dat.loc['quantile', param]= bootstrap_quantile_confidence_interval(np.mean(df[param]), est)
    dat.loc['interval_length', param]= abs(bootstrap_normal_confidence_interval(np.mean(df[param]), est)[0]-bootstrap_normal_confidence_interval(np.mean(df[param]), est)[1])
```

```python
def draw_confidence_interval(
    left,
    right,
    estimation=None,
    sample=None, 
    ylim=(None, None), 
    estim_label = '',
    sample_label='',
    color=None,
    interval_label=None
):
    '''
    Рисует доверительный интервал и оценку в зависимости от размера выборки.
    
    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param estim_label: подпись к оценке
    :param sample_label: подпись к выборке
    :param color: цвет, которым будет отображен доверительный интервал
    '''
    plt.figure(figsize=(15,7))
    
    n_range = np.arange(len(estimation))
    
    plt.plot(n_range+1, left, color=color, label = interval_label)
    plt.plot(n_range+1, right, color=color)

    plt.plot(n_range + 1 , estimation, color = 'red', label= estim_label)
    
    plt.xlabel('Размер выборки')
    plt.ylabel('Доверительный интервал')
    plt.ylim(ylim)
    plt.legend()
    plt.title(sample_label)
    plt.show()
```

**Решение:**

```python
# Нормальный доверительный интервал
for param in params:
    sample = df[param].sample(n=100, random_state=42)
    n_range = np.arange(1, len(sample) + 1)
    lefts = []
    rights = []
    estimation = []
    for i in range(1, len(sample) + 1):
        left, right = bootstrap_normal_confidence_interval(np.mean(sample[:i]), bootstrap(sample[:i]))
        lefts.append(left)
        rights.append(right)
        estimation.append(np.mean(sample[:i]))
    draw_confidence_interval(lefts, rights, estimation, sample, ylim=(min(sample), max(sample)), estim_label='Оценка', sample_label = param, color='blue', interval_label='Доверительный интервал')    
    
```
